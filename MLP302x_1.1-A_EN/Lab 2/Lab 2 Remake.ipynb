```
import numpy as np
import pandas as pd
import sklearn
```

```
df = pd.read_csv("/content/drive/MyDrive/FUNIX Progress/MLP302x_1.1-A_EN/data/kc_house_data.csv", index_col=0)
```

```
df.head()
```

```
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, train_size = .8, test_size = .2, random_state = 0)
```

```
def feature_extraction(data, feature_list):
  return data[feature_list].values
```

```
from sklearn.linear_model import LinearRegression
example_features_list = ['sqft_living', 'bedrooms', 'bathrooms']
example_features = feature_extraction(train, example_features_list)
example_labels = train['price']
example_model = LinearRegression().fit(example_features, example_labels)
```

```
example_model.coef_
```

```
example_model.predict(example_features)[0]
```

```
def rss(model, data, label):
  return sum((label - model.predict(data))**2)
```

```
rss_test = rss(example_model, feature_extraction(test, example_features_list), test['price'])
print(rss_test)
```
